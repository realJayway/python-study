# Mutiplication Pictures

```python
t=linspace(0,2*pi,50)
x=sin(t)
y=cos(t)
figure()
plot(x)
figure()
plot(y)
```

```python
subplot(1,2,1)
plot(x)
subplot(1,2,2)
plot(y)
```

# Labels

```python
plot(x,label='sin')
plot(y,label='cos')
legend()
```

Or we can insert `legend` directly:

```python
plot(x)
plot(y)
legend(['sin','cos'])
```

# Coordinate Axises, Titles, Girds

```python
plot(x,sin(x))
xlabel('radians')
# set figures' sizes
ylabel('amplitude', fontsize='large')
title('sin(x)')
# show the gird
grid()
```
