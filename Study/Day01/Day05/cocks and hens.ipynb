```python
"""
鸡翁一值钱五，鸡母一值钱三，鸡雏三值钱一。百钱买百鸡，问鸡翁、鸡母、鸡雏各几何？
"""

for x in range(0,20):
    for y in range(0,(100-5*x)//3+1):
        for z in range(0,100-5*x-3*y+1): #错误：剩下的所有钱都要用来买小鸡
            if x+y+3*z==100:
                print('鸡翁有%d只，鸡母有%d只，鸡雏有%d只'%(x,y,3*z))
```

```python
for x in range(0,20):
    for y in range(0,(100-5*x)//3+1):
            if x+y+3*(100-5*x-3*y)==100:
                print('鸡翁有%d只，鸡母有%d只，鸡雏有%d只'%(x,y,100-x-y))
```
