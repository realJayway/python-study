## 折线图的基础属性设置

基础属性设置  
```
color = 'red'    : 颜色
alpha = 0.5      : 透明度
linestyle = '--' : 样式
linewidth = 3    : 宽度

```

````python
# 折线图
from matplotlib import pyplot as plt

x = range(1,8)
y = [17, 17, 18, 15, 11, 11, 13]

# 传入x和y，通过plot画折线图
plt.plot(x,y,color = 'red', alpha = 0.5, linestyle = '--', linewidth = 3) # alpha表示透明度可取[0, 1]
plt.show() # show进行展示，好像不用也可以？ 只是少了一行图片的模块
````

```
- 实线(solid)
-- 短线(dashed)
-. 断点相间线(dashdot)
:  虚点线(dotted)
```

## 绘制x,y轴

````python
from matplotlib import pyplot as plt
import random

x = range(2,26,2)
y = [random.randint(15,30) for i in x]
plt.figure(figsize = (20,8), dpi = 80)

# 设置x轴的刻度的疏密程度
# plt.xticks(x)
# plt.xticks(range(1,25)) 
# 上面两种均可以

# 修改x轴刻度
x_tick_label = ["{}:00".format(i) for i in x]
plt.xticks(x,x_tick_label)

# 设置y轴
y_tick_label = ["{}℃".format(i) for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y)), y_tick_label)



plt.plot(x,y)
plt.show()
````

## 添加标题及设置显示中文

matplotlib只显示英文，我们需要通过font_manager添加字体解决该问题

````python
from matplotlib import pyplot as plt
import matplotlib
import random

x = range(120)
y = [random.randint(10,30) for i in range(120)]

plt.figure(figsize = (20,8), dpi = 80)
plt.plot(x,y)

from matplotlib import font_manager
# 加载系统字体
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\simhei.ttf',size=36)
# 设置图片标题
plt.title('图片',fontproperties=my_font,color = 'red') # 直接 使用中文会报错：缺少字体
plt.xlabel('时间',fontproperties=my_font, rotation=45)
plt.show()
````

````python
a = ['流浪地球','疯狂的外星人','飞驰人生','大黄蜂','熊出没·原始时代','新喜剧之王']
b = [38.13,19.85,14.89,11.36,6.47,5.93]

plt.figure(figsize = (20,8), dpi = 80)
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\simhei.ttf',size = 10)

rects = plt.bar(range(len(a)),[float(i) for i in b], width = 0.3, color = ['r', 'g', 'b', 'r', 'g', 'b'])
plt.xticks(range(len(a)), a, fontproperties = my_font)
plt.yticks(range(0,41,5),range(0,41,5))

# 在条形图上加数据值
for rect in rects:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height + 0.3, str(height), ha = "center")  # 横向坐标，纵向坐标，字符串，模式

plt.show()
````
